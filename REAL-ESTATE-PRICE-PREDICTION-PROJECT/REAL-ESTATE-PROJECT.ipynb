```python
training_set=strat_train_set.drop("MEDV",axis=1)
labels_set=strat_train_set['MEDV'].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(training_set)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(training_set)
```

```python
housing_tr = pd.DataFrame(X, columns=training_set.columns)
```

```python
housing_tr.describe()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
        ('std_scaler', StandardScaler()),
])
```

```python
df_training=my_pipeline.fit_transform(housing_tr)
```

```python
df_training
```

```python
df_training.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(df_training, labels_set)
```

```python
somedata=housing_tr.iloc[:5]
somelabels=labels_set.iloc[:5]
```

```python
predicting_dataset=my_pipeline.transform(somedata)
```

```python
model.predict(predicting_dataset)
```

```python
list(somelabels)
```

```python
from sklearn.metrics import mean_squared_error
housing_preds=model.predict(df_training)
mse=mean_squared_error(labels_set,housing_preds)
mse
```

```python
import numpy as np
rmse=np.sqrt(mse)
```

```python
rmse
```

```python
X_test = strat_test_set.drop("MEDV", axis=1)
Y_test = strat_test_set["MEDV"].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```
